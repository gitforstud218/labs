### Державний устрій

    Йорданія є конституційною монархією, де головну роль у державі відіграє король. Нинішній король Йорданії — **Абдалла II**, представник Хашимітської династії, яка бере свій початок від самого пророка Мухаммеда. Хашиміти правлять Йорданією протягом багатьох десятиліть, і король Абдалла II продовжує цю традицію. Політична система поєднує в собі елементи монархії та парламентаризму. Король має значні повноваження, однак є й парламент, який грає важливу роль у прийнятті законів та формуванні уряду. Йорданія є однією з небагатьох арабських держав, де існує реальний політичний плюралізм і свобода вираження думок, хоча ця свобода має свої межі.
    Королева **Ранія**, дружина короля Абдалли II, відома своєю активною гуманітарною діяльністю. Вона займається питаннями освіти, прав жінок і дітей, а також активно виступає за соціальні реформи в Йорданії та в арабському світі загалом.

## Визначні постаті Йорданії

- **Король Абдалла II** — нинішній монарх Йорданії, що править з 1999 року. Він відіграє важливу роль у міжнародній політиці та стабільності регіону.
- **Королева Ранія** — відома активістка, що займається питаннями освіти, соціальної рівності та прав людини.
- **Лоуренс Аравійський** — легендарний британський офіцер, що активно співпрацював з арабськими повстанцями під час Першої світової війни.

## Культурні та природні пам'ятки

Йорданія має багату культурну спадщину, яка приваблює туристів і дослідників з усього світу. Тут можна знайти безліч історичних пам'яток, які є свідченням древньої історії цього регіону. Крім того, Йорданія має унікальні природні ландшафти, що створюють неповторний колорит цієї країни.

### Петра

*Петра* — це одне з найвідоміших стародавніх міст світу, вирубане у скелях. Воно було засноване приблизно у 312 році до н.е. і стало столицею Набатейського царства. Місто Петра має багату архітектурну спадщину, що включає храми, гробниці та театри. Його внесено до списку Всесвітньої спадщини ЮНЕСКО, і воно є однією з найпопулярніших туристичних атракцій в Йорданії.
- Столиця Набатейського царства
- Вирубане в скелях
- Одне з нових семи чудес світу

### Мертве море

*Мертве море* — це унікальне природне явище, відоме своїм надзвичайно високим вмістом солі. Воно розташоване на кордоні між Йорданією та Ізраїлем і є найсолонішим водоймищем на Землі. Завдяки своїм хімічним властивостям, вода в Мертвому морі дозволяє людині буквально плавати на поверхні, що є унікальним досвідом для туристів.
- Найсолоніше озеро на планеті
- Розташоване на кордоні з Ізраїлем
- Відоме своїми цілющими властивостями

### Ваді Рам

*Ваді Рам*, відоме також як Місячна долина, — це пустеля, яка вражає своїми фантастичними ландшафтами. Тут можна побачити величезні скелі, що піднімаються над червоним піском, і неймовірні панорами, які виглядають як декорації для науково-фантастичного фільму. Ця місцевість є важливим туристичним напрямком та має значення для національної ідентичності Йорданії.
- Пустельний регіон на півдні Йорданії
- Відомий своїми червоними пісками та скелями
- Популярне місце для туристів та кінозйомок

```python
def task(first, second):
    arr = []  
    for i in range(first, second): 
        if i > 1: 
            prime = True 
            for j in range(2, int(i ** 0.5) + 1):
                if i % j == 0: 
                    prime = False 
            if prime: 
                arr.append(i)  
    print(arr)  

task(3, 15) 
task(6, 24) 
task(7, 12)
```
